# Map code for presentation

```python
# import libraries
import folium
import pandas as pd
 
# Make a data frame with dots to show on the map
data = pd.DataFrame({
'lat':[55.75, 59.57, 55.78, 56.50, 51.40, 48.71, 56.32, 54.98, 54.73],
'lon':[37.61, 30.19, 49.12, 60.35, 39.11, 44.50, 44.00, 73.36, 55.95],
'name':['Moscow', 'St. Petersburg', 'Kazan', 'Yekaterinburg', 'Voronezh', 'Volgograd', 'Nizhniy Novgorod', 'Omsk', 'Ufa']
})
data
 
# Make an empty map
m = folium.Map(location=[52.23, 54.58], tiles="cartodbpositron", zoom_start=3)


# I can add marker one by one on the map
for i in range(0,len(data)):
    folium.Marker([data.iloc[i]['lat'], data.iloc[i]['lon']], popup=data.iloc[i]['name'], icon=folium.Icon(color='darkred')).add_to(m)
m
```
